```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import fpgrowth
```

# Dataset 1

```python
data = pd.read_csv('groceries - groceries.csv')
```

```python
data.head()
```

```python
data.drop('Item(s)',axis=1,inplace=True)
```

```python
data.shape
```

```python
list_of_items=[]
for i in data:
    for j in data[i]:
        j=str(j)
        if j != 'nan':
            list_of_items.append(j)
```

```python
list_of_items= sorted(set(list_of_items))
```

```python
list_of_items
```

```python
formatted_data = pd.DataFrame(np.zeros((9835,169))==1,columns=list_of_items)
```

```python
for idx, row in data.iterrows():
    for item in row:
        for i in formatted_data:
            if i==item:
                formatted_data[i][idx]=True
```

```python
formatted_data.head()
```

```python
output_ap = apriori(formatted_data, min_support = 0.05, use_colnames = True)
output_ap
```

```python
output_fp = fpgrowth(formatted_data, min_support = 0.05, use_colnames = True)
output_fp
```

```python
output_ap = apriori(formatted_data, min_support = 0.01, use_colnames = True)
output_ap
```

```python
output_fp = fpgrowth(formatted_data, min_support = 0.01, use_colnames = True)
output_fp
```

```python
all_items
```
